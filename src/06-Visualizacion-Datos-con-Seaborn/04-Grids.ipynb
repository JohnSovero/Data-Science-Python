# Grids

Las cuadrículas son tipos generales de gráficos que permiten mapear tipos de visualizaciones en filas y columnas de una cuadrícula, esto ayuda a crear gráficas similares separadas por características.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

## PairGrid

**Pairgrid** es una cuadrícula de subplot para graficar relaciones por pares en un conjunto de datos.

```python
# Solo la cuadrícula
sns.PairGrid(iris)
```

```python
# Luego mapea a la cuadrícula
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
# Asignar a superior, inferior y diagonal
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

## pairplot

**pairplot** es una versión más simple de PairGrid (se usa con bastante frecuencia)

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species',palette='rainbow')
```

## Facet Grid

**FacetGrid** es la forma general de crear cuadrículas de gráficos basados en una función:

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Solo la cuadrícula
g = sns.FacetGrid(tips, col="time", row="smoker")
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill")
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker",hue='sex')
# Observa cómo los argumentos vienen después de la llamada plt.scatter
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```

## JointGrid

**JointGrid** es la versión general para cuadrículas de tipo jointplot().  Un ejemplo rápido:

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot(sns.regplot, sns.distplot)
```

Consulta la documentación según sea necesario para los tipos de cuadrícula, pero la mayoría de las veces solo se usan los gráficos más fáciles discutidos anteriormente.
