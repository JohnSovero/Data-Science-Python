# Gráficos de regresión

Seaborn tiene muchas capacidades integradas para gráficos de regresión, sin embargo, solo cubriremos la función **lmplot()**.

**lmplot** permite mostrar modelos lineales, pero también convenientemente permite dividir esos gráficos en función de las características, así como colorear el tono en función de las características.

Recordar que el análisis de la regresión lineal se utiliza para predecir el valor de una variable según el valor de otra. La variable que desea predecir se denomina variable dependiente. La variable que está utilizando para predecir el valor de la otra variable se denomina variable independiente.

Exploremos cómo funciona esto:

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## lmplot()

```python
sns.lmplot(x='total_bill',y='tip',data=tips, fit_reg=False) 
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

## Usando una cuadrícula (Grid)

Podemos agregar más separación de variables a través de columnas y filas con el uso de una cuadrícula. Simplemente indicamos esto con los argumentos col o row:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm')
```

## Aspecto y tamaño 

Las figuras de Seaborn pueden ajustar su tamaño y relación de aspecto con los parámetros **size** y **aspect**:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm',
          aspect=0.6, height=8)
```
