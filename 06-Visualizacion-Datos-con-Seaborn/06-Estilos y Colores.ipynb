# Estilo y color

Repasemos cómo controlar la estética de las figuras en seaborn:

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
```

## Estilos

Puede establecer estilos particulares:

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('dark') # valores posibles: white, dark, whitegrid, darkgrid, ticks
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips, palette='deep')
```

## Eliminación de columna

```python
sns.countplot(x='sex', data=tips)
sns.despine()
```

```python
sns.countplot(x='sex', data=tips)
sns.despine(left=True)
```

## Tamaño y aspecto

Se puede usar **plt.figure(figsize=(width,height)** de matplotlib para cambiar el tamaño de la mayoría de diagramas seaborn.

Se puede controlar el tamaño y la relación de aspecto de la mayoría de los diagramas de cuadrícula de Seaborn pasando parámetros: size y aspect. Por ejemplo:

```python
# Gráfico sin cuadrícula
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
# Gráfico de tipo de cuadrícula
sns.lmplot(x='total_bill',y='tip', height=2,aspect=4,data=tips)
```

## Escala y contexto

La funcion set_context() permite anular los parámetros predeterminados:

```python
sns.set_context('poster',font_scale=4)
sns.countplot(x='sex',data=tips,palette='coolwarm')
```
